```python
import numpy as np
import pandas as pd
```

```python
col = ['userid', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=col)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
movie_titles.head()
```

```python
mer=pd.merge(df,movie_titles,on='item_id')
```

```python
mer
```

```python
mer.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
mer.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rat=pd.DataFrame(mer.groupby('title')['rating'].mean())
rat.head()
```

```python
rat['count_ratings']=pd.DataFrame(mer.groupby('title')['rating'].count())
```

```python
rat.sort_values('count_ratings',ascending=False)
```

```python
mov=mer.pivot_table(values='rating',index='userid',columns='title')
```

```python
mov
```

```python
rat.sort_values('count_ratings',ascending=False)
```

```python
star=mov['Star Wars (1977)']
far=mov['Fargo (1996)']
```

```python
star.dropna(inplace=True)
far.dropna(inplace=True)
```

```python
star=star.join(rat['count_ratings'])
far=far.join(rat['count_ratings'])
```

```python
simstar=mov.corrwith(star)
simfar=mov.corrwith(far)
```

```python
simfar.dropna(inplace=True)
simstar.dropna(inplace=True)
```

```python
simfar=pd.DataFrame(simfar,columns=['corr'])
simstar=pd.DataFrame(simstar,columns=['corr'])
```

```python
simfar=simfar.join(rat['count_ratings'])
```

```python
simstar=simstar.join(rat['count_ratings'])
```

```python
sf=simfar[simfar['count_ratings']>100].sort_values('corr',ascending=False).drop(['count_ratings','corr'],axis=1).head(10)
```

```python
ss=simstar[simstar['count_ratings']>100].sort_values('corr',ascending=False).drop(['count_ratings','corr'],axis=1).head(10)
```

```python
print('movies recommended according to fargo :-')
for i in sf.index:
    print(i)
    
```

```python
print('movies recommended according to Star Wars:-')
for i in ss.index:
    print(i)
    
```
